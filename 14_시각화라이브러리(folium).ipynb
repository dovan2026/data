# 4.시각화 라이브러리 : folium

- https://python-visualization.github.io/folium/latest/index.html

- **지도(Maps)** 를 활용해 공간 데이터를 시각화할 수 있는 Python 라이브러리

- 인터랙티브 지도를 쉽게 만들 수 있어서 공간 데이터 분석, 경로 시각화, 위치 데이터 분석 등에 많이 사용됨

````python
# %pip install folium

import folium
````

````python
loc = [37.4864987317089,127.020663860591]

f = folium.Figure(width=700, height=500)
m = folium.Map(location=loc, zoom_start=17).add_to(f)
m
````

````python
m.save('test2.html')
````

## **4.1. 특정 장소의 지도 시각화하기**
- 네이버 지도에서 원하는 위치를 검색하고 공유 버튼을 통해 URL 복사
- [링크](https://xn--yq5bk9r.com/blog/map-coordinates)에 복사한 url을 붙여넣고 위도와 경도 확인
- 위도 경도 찾아주는 다른 사이트 : [링크](https://www.findlatlng.org/)

- 기본 문법

```
f = folium.Figure(width=가로길이, height=세로길이)
m = folium.Map(location=[위도, 경도], zoom_start=줌할정도).add_to(f)
m.save('test.html') #지도 저장
m
```


````python
folium.Marker(loc, tooltip='보금자리').add_to(m)
m
````

````python
folium.Marker(loc,
              tooltip='보금자리',
              icon=folium.Icon(color='red', icon='star')).add_to(m)
m
````

## **4.2. 마커 추가하기**

- 장소 표시 마커


```
folium.Marker([위도, 경도]
                , tooltip=마우스 오버시 나타남
                , popup=클릭시 나타남
                , icon=folium.Icon(color=색, icon=모양)).add_to(지도)
```



- 원 형태 마커

```
folium.CircleMarker([위도, 경도]
                , radius=범위
                , color=색).add_to(지도)
```



````python
loc = [37.5511694, 126.9882266] 

f = folium.Figure(width=600, height=400)
m = folium.Map(location=loc,
               zoom_start=15).add_to(f)

folium.Marker(loc,
              tooltip='남산서울타워',
              icon=folium.Icon(color='blue', icon='star'),
              popup ='<iframe src="https://i.pinimg.com/736x/51/dc/c6/51dcc6c4dcff66b1008d37d752a42d2a.jpg"></iframe>'
              ).add_to(m)
m
````

````python
folium.CircleMarker(
    loc,
    color='red',
    radius=100,
).add_to(m)
m
````
